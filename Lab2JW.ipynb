```python
# Find desired sequences in cluster for bar plot
clusters = []
cluster1 = ['hu.39','rh.50','rh.57','rh.51','rh.49','rh.53','rh.64','rh.61','rh.58']
cluster2 = ['pi.2','pi.3','pi.1']
cluster3 = ['rh.10','bb.1','bb.2','hu.17','hu.6']
cluster4 = ['rh.2','rh.40','hu.67','hu.37','hu.40','hu.66','hu.41','rh.38','hu.42']
cluster5 = ['cy.3','cy.6','cy.4','rh.13']
cluster6 = ['rh.35','rh.36','rh.37']
cluster7 = ['hu.43','hu.48','hu.44','hu.46']
cluster8 = ['hu.14','hu.31','hu.31','rh.43']
clusters = [cluster1, cluster2, cluster3, cluster4, cluster5, cluster6, cluster7, cluster8]
num_clusters = len(clusters)

# Get clusters' sequences
seq_dict = {k: [] for k in range(num_clusters)}
for sequence in SeqIO.parse('seqs.aligned.fa', 'fasta'):
    for c in range(len(clusters)):
        if sequence.id in clusters[c]:
            seq_dict[c].append(sequence.seq.upper())
```

```python
# Input plot range
min_pos = 0
max_pos = 2000
plot_range = range(min_pos, max_pos)

fig = plt.figure()
fig.set_figheight(4*num_clusters)
fig.set_figwidth(200)

for c in range(num_clusters):
    ax = fig.add_subplot(num_clusters,1,c+1)
    seqs = seq_dict[c]
    # Initialize dictionary of counts
    content_dict = {k: [0,0,0] for k in plot_range}  # [A/T count, G/C count, indel count]
    # Get counts
    for seq in seqs:
        for pos in plot_range:
            if seq[pos] == 'A' or seq[pos] == 'T':
                content_dict[pos][0] += 1
            elif seq[pos] == 'C' or seq[pos] == 'G':
                content_dict[pos][1] += 1
            else:
                content_dict[pos][2] += 1

    # Normalize
    num_seqs = len(seqs)
    for pos, counts in content_dict.items():
        content_dict[pos] = [count/num_seqs for count in counts]

    # Plotting bar plot of MSA
    counts = sorted(content_dict.items())
    positions, counts = zip(*counts)
 
    AT_counts = [count[0] for count in counts]
    GC_counts = [count[1] for count in counts]
    ATGC_bottom = [count[0] + count[1] for count in counts]
    indel_counts = [count[2] for count in counts]

    pAT = ax.bar(positions, AT_counts, color = 'r')
    pCG = ax.bar(positions, GC_counts, bottom = AT_counts, color = 'b')
    pND = ax.bar(positions, indel_counts, bottom = ATGC_bottom, color = 'y')
    
    plt.title('Cluster ' + str(c+1))
    plt.xlim([min_pos, max_pos])
    plt.xlabel('Position in sequence')
    plt.ylabel('%AT or %CG')
    plt.xticks(np.arange(min_pos, max_pos, 50))
    plt.subplots_adjust(hspace=0.5)
    plt.legend([pAT,pCG,pND], ['AT','CG','indel'], loc=2)

plt.show()   # double-click to enlarge
```

```python
# Find desired clusters' sequence lengths
lens_dict = {k: [] for k in range(num_clusters)}    # initialize dict of clusters:sequence lengths
for sequence in SeqIO.parse('seqs.fa', 'fasta'):
    for cluster in range(num_clusters):
        if sequence.id in clusters[cluster]:
            lens_dict[cluster].append(len(sequence.seq))
```

```python
# Plotting box plot of sequence lengths
all_lens = []
xlabels = []
for cluster, lens in lens_dict.items():
    all_lens.append(lens)
    xlabels.append('Clust ' + str(cluster+1))

fig, ax = plt.subplots(figsize=(30,10))
plt.boxplot(all_lens)
plt.xlabel('Clusters')
plt.ylabel('Sequence Length')
ax.set_xticklabels(xlabels)
ax.grid()
plt.show()   # double-click to enlarge
```
